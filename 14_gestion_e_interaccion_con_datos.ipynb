## Obtención de un conjunto de datos mediante los métodos *data()* y *enter()* .

```python
%%html
<div id="contenedor-1"></div>
```

```python
%%javascript

const datos_1 = [1, 2, 3, 4, 12 ,17];
require(["d3"], function(d3) { 
    d3.select("#contenedor-1").
        selectAll("p").
        data(datos_1).
        enter().
        append("p").
        text(d => d);
});
```

```python
%%html
<div id="contenedor-2">
</div>
```

```python
%%javascript

const datos_2 = [1, 2, 3, 4, 12, 3, 1, 15];

require(["d3"], function(d3) {
    d3.select("#contenedor-2").
        selectAll("div").
        data(datos_2).
        enter().append("div").
        style("background-color", "blue").
        style("color", "white").
        style("text-align", "center").      
        style("width", d => d * 10 + "px").
        text(d => d);
        })
```

```python
%%html
<div id="contenedor-3">
<svg id="svg-3" width="100" height="70">
</div>
```

```python
%%javascript

const datos_3 = [1, 2, 3, 2.3, 6, "hola", 12];

let contador = -15;

function suma(){
    contador += 15;
    return contador;
}

require(["d3"], function(d3) {
    d3.select("#svg-3").
        selectAll("rect").
        data(datos_3).
        enter().append("rect").
        attr("fill", "black").
        attr("x", "0").
        attr("height", "10").
        attr("width", d => d * 10).
        attr("y", d => suma());
        })
```

```python
%%javascript

const matriz = [
  [11975,  5871, 8916, 2868],
  [ 1951, 10048, 2060, 6171],
  [ 8010, 16145, 8090, 8045],
  [ 1013,   990,  940, 6907]
];

require(["d3"], function(d3) { 
    let tr = d3.select("#contenedor-4")
        .append("table")
        .selectAll("tr")
        .data(matriz)
        .enter().append("tr");
    
    let td = tr.selectAll("td")
        .data(d => d)
        .enter().append("td")
        .text(d => d);
});
```

### Carga de documentos de texto.

```python
%%html
<div id="contenedor-5">
</div>
```

```python
%%javascript

require(["d3"], function(d3){
    d3.text("data/texto.txt",).then(function(datos){
            d3.select("#contenedor-5").
            append("p").
            text(datos);
    })
});
```

### Carga de imágenes.

```python
%%html
<div id="contenedor-6">
</div>
```

```python
%%javascript

require(["d3"], function(d3) {
    d3.image("imagenes/pythonista.png").then(function(imagen) {
        let contenedor = document.getElementById("contenedor-6");
        contenedor.appendChild(imagen);
    });
});
```

### Carga de documentos HTML.

```python
%%html
<div id="contenedor-7">
</div>
```

```python
%%javascript

require(["d3"], function(d3) {
    d3.html("data/codigo.html").then(function(codigo){
        let contenedor = document.getElementByID("#contenedor-7");
        contenedor.text(codigo.textContent);
    });
});
```

### Carga de  de documentos CSV.

**Ejemplo:**

Se leerá el archivo [data/calificaciones.csv](data/calificaciones.csv) y se desplegarán las propiedades:
* *Nombre*
* *py101*
* *py111*
* *py121* 

```python
%%html
<div id="contenedor-8">
</div>
```

```python
%%javascript

require(["d3"], function(d3) {
    d3.csv("data/calificaciones.csv",).then(function(datos){
        d3.select("#contenedor-8").
            selectAll("p").
            data(datos).
            enter().append("p").
            text(d => d.Nombre + ". py101: " + d.py101 + ". py111: " + d.py111 
                 + ". py121: " + d.py121);
    })
});
```
